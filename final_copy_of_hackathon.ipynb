```python
import pandas as pd
import os


def main():
    """Main function to handle user interaction and data processing."""
    data_source = input("Do you want to enter patient data manually or use a CSV file? (manual/csv): ").lower()

    if data_source == "manual":
        num_patients = int(input("Enter the number of patients: "))
        patients_data = []
        for _ in range(num_patients):
            name = input("Enter patient's name: ")
            age = int(input("Enter patient's age: "))
            address = input("Enter patient's address: ")
            patients_data.append({"Name": name, "Age": age, "Address": address})
        original_filename = "original_data_manual.csv"
        pd.DataFrame(patients_data).to_csv(original_filename, index=False)
        print(f"\nOriginal data saved to '{original_filename}'")
    elif data_source == "csv":
        csv_file = input("Enter the path to the CSV file: ")
        try:
            patients_data = pd.read_csv(csv_file).to_dict(orient="records")
            original_filename = "original_data.csv"
            pd.DataFrame(patients_data).to_csv(original_filename, index=False)
            print(f"\nOriginal data saved to '{original_filename}'")
        except FileNotFoundError:
            print("File not found. Please provide a valid path.")
            return
    else:
        print("Invalid choice. Please enter 'manual' or 'csv'.")
        return


    print("\nNow you can proceed with encryption or decryption steps.")

if __name__ == "__main__":
    main()
```

```python
import hashlib
import json
import pandas as pd
import os
import getpass

def decrypt_data(encrypted_data, key):
    """Decrypt data using provided key."""
    decrypted_data = []
    for hashed_item in encrypted_data:
        decrypted_data.append(encrypted_data[hashed_item])
    return decrypted_data

def get_user_input():
    """Get user input for decryption."""
    key = getpass.getpass("Enter decryption key: ")
    return key

def save_to_csv(data, filename):
    """Save decrypted data to CSV file."""
    df = pd.DataFrame(data)
    df.to_csv(filename, index=False)
    print(f"Decrypted data saved to '{filename}'")

def main():

    encrypted_data_file = input("Enter the path to the encrypted CSV file: ")
    try:
        encrypted_df = pd.read_csv(encrypted_data_file)
        encrypted_data = encrypted_df.to_dict(orient="index")
    except FileNotFoundError:
        print("File not found. Please provide a valid path.")
        return


    decryption_key = get_user_input()


    decrypted_data = decrypt_data(encrypted_data, decryption_key)


    output_filename = "decrypted_data1.csv"
    save_to_csv(decrypted_data, output_filename)


    os.system(f'start explorer /select,"{os.path.abspath(output_filename)}"')

if __name__ == "__main__":
    main()
```
